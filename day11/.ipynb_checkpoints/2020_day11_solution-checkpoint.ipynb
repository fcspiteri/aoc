```python
with open('input11.txt','r') as f_open:
    data = f_open.read()
data = data.split('\n')

rows = len(data)
cols = len(data[0])

from collections import Counter
import copy
```

```python
def swap_seats(lst):
    changes = 0
    data_new = lst.copy()
    for row in range(0, rows):
        for col in range(0, cols):
            adjacent = []
            for i in range(-1, 2):
                for j in range(-1, 2):
                    try:
                        if (0 <= row+i <= rows) and 0 <= col+i <= cols:
                            adjacent.append(lst[row+i][row+j])
                    except: 
                        continue
            if lst[row][col] == "L":
                if "#" not in adjacent:
                    string = list(data_new[row])
                    string[col] = "#"
                    data_new[row] = "".join(string)
                    changes = changes + 1
            if lst[row][col] == "#":
                if Counter(adjacent)["#"] >= 4:
                    string = list(data_new[row])
                    string[col] = "#"
                    data_new[row] = "".join(string)
                    changes = changes + 1
            #print(row, col, lst[row][col], data_new[row][col], adjacent)
    occupied = 0
    for row in data_new:
        occupied += row.count("#")
    print(changes, occupied)
    if changes > 0:
        swap_seats(data_new)
    return changes, occupied, data_new


print(swap_seats(data))
```

```python

seat_swaps = 1
while True:
    if seat_swaps == 0:
        break
    else:
        seat_swaps, occupied, new_data = swap_seats(data)
        print(seat_swaps, occupied)
        data = new_data.copy()
```
